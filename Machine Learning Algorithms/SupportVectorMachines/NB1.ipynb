## Simple Support Vector Machine in action
Consider the case of two seperable classes for the display of SVM. The two classes are well separated.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# use seaborn plotting defaults
import seaborn as sns; sns.set()
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=50, centers=2,
                  random_state=0, cluster_std=0.60)
plt.figure(figsize=(20,10))
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn');
```

### Case of Best Linear Discriminator
There are tons of best fit lines that can discriminate the two classes. The challenge is to find the best fit line that can explain the variation between the classes. Consider the case demonstrated next on the number of possible solution when drawn by hand.

```python
xfit = np.linspace(-1, 3.5)
plt.figure(figsize=(20,10))
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
plt.plot([0.6], [2.1], 'x', color='red', markeredgewidth=2, markersize=10)

for m, b in [(1, 0.65), (0.5, 1.6), (-0.2, 2.9)]:
    plt.plot(xfit, m * xfit + b, '-k')

plt.title('Best Discriminators chosen at Random - CHALLENGE!')
plt.xlim(-1, 3.5);
```

Based on the line considered as the solution the value $X$ will be assigned either the red or the yellow class. But which line can be the best to go with?
