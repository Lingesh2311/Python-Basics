### PCA for Visualization - Handwritten Digits

This can be useful to get a lower dimensional view of the handwritten MNIST digits dataset. The images are stored as $64$ dimensional arrays. We can visualize them in the lower dimension (2) to get a better insight on how the data varies when considering the first 2 principal components.

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits

digits = load_digits()
print(digits.data.shape)
```

```python
pca = PCA(2) # To convert 64 dimension to 2 dimension data
projected_digits = pca.fit_transform(digits.data)
print(f'Original Data Shape: {digits.data.shape}')
print(f'Transformed Data Shape: {projected_digits.data.shape}')
```

```python
plt.figure(figsize=(10,10))
plt.scatter(projected_digits[:,0], projected_digits[:,1],
            edgecolors='none', alpha=0.5,
            cmap=plt.cm.get_cmap('rainbow', 10), c=digits.target)
plt.xlabel('Component 1'); plt.ylabel('Component 2'); plt.colorbar()
_ = plt.title('Projection of 64 dimensions in a plane after PCA', fontsize=15)
```
