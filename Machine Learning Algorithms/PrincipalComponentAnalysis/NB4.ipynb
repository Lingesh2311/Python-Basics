### Choosing the number of components

This is a vital part of the PCA decomposition. The number of components required for replicating the dataset without much loss in information can be found by plotting the cumulative sum of explained variance ratio as the number of components vary. 

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_digits

digits = load_digits()
pca = PCA()
pca = pca.fit(digits.data)
plt.figure(figsize=(10,10))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components'); plt.ylabel('Cumulative Sum of Variance ratio')
_ = plt.title('Variance Ratio vs. number of components')
```

We see that we need about 20 components to get the explaination of 90% of the variance.
