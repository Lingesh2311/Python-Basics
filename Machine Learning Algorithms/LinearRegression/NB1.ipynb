## General Linear Regression

This is the simplest of them all. A straight line fit for the data.
<br>
$y = ax + b$
<br>
The `sklearn.linear_model.LinearRegression` model can be used to fit lines, planes, and hyperplanes to fit higher dimensional data.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
```

```python
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
# Consider the line with slope 2 and intercept -5
y = 2 * x - 5 + rng.randn(50)
plt.figure(figsize=(10,10))
plt.scatter(x, y);
```

Use the `sklearn.linear_model.LinearRegression` tool for fitting a straight line

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)

model.fit(x[:, np.newaxis], y)

xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.figure(figsize=(10,10))
plt.scatter(x, y)
plt.title('Straight Line fit using the Linear Regression Model')
plt.plot(xfit, yfit);
```

```python
# Slope and Intercept Attributes can be extracted
print("Model slope:    ", model.coef_[0])
print("Model intercept:", model.intercept_)
```
