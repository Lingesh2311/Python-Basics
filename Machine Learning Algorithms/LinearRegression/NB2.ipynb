## Polynomial Basis Function

This is done to fit features ($x_{i}$) that have non-linear relationship with the target variable ($y$). This is done using the `sklearn.preprocessing.PolynomialFeatures` model followed by a `fit_transform()` method. This will create more columns based on the degree of the polynomial specified.

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np
x = np.array([2, 3, 4])
poly = PolynomialFeatures(3, include_bias=False)
poly.fit_transform(x[:, None])
```

```python
from sklearn.pipeline import make_pipeline
poly_model = make_pipeline(PolynomialFeatures(7), LinearRegression())
```

```python
import matplotlib.pyplot as plt

rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = np.sin(x) + 0.1 * rng.randn(50)

poly_model.fit(x[:, np.newaxis], y)
xfit = np.linspace(0, 10, 1000)
yfit = poly_model.predict(xfit[:, np.newaxis])

plt.figure(figsize=(20,10))
plt.scatter(x, y)
plt.xticks([]); plt.yticks([])
plt.plot(xfit, yfit);
```
